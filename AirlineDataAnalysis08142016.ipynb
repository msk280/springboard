<h1 id="Analysis-of-Airline-Security Delays"><span style="color: #ff0000;">Analysis of Airline Security Delays</span></h1>

Many of us will already know this. Security delay is caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas. In this project, we analyze security delay that happens during Air Travel using exploratory analysis and also develop machine learning models to predict future security delays.

<p>The Security Delay focuses on 2014 Top 10 Airports. We need to look at all the available attributes in the data set and pick the variables that are potential candidates for the model.</p>
<p>After carefully checking the attributes, we pick the following:-</p>
<p>1. Month</p>
<p>2. Day of month</p>
<p>3. Day of week</p>
<p>4. Departure Hour</p>

Preparing data - The data will be trimmed to contain only the required information for our analysis.

```python
# Trimming the data set needed for this analysis
Temp1 = pd.read_csv("Analysis2014.csv")
Temp2 = Temp1[Temp1.SECURITY_DELAY > 0]
Temp3 = Temp2.loc[:,('MONTH','DAY_OF_MONTH','DAY_OF_WEEK','DEP_HOUR','SECURITY_DELAY','SECURITY_DELAY_FLAG' )]
Temp4 = Temp3.groupby(['MONTH','DAY_OF_MONTH','DAY_OF_WEEK','DEP_HOUR']).SECURITY_DELAY.sum()
Temp5 = pd.DataFrame({'SECURITY_DELAY' : Temp4})
Temp5.index.names = ['MONTH','DAY_OF_MONTH','DAY_OF_WEEK','DEP_HOUR']
Temp5.to_csv('Temp5.csv') 
```

Feature binarization is the process of thresholding numerical features to get boolean values.
For the purpose of this model, we will use feature binarization on Month, Day of Month, Day of Week and Departure Hour.

```python
array = Temp5.values
# separate array into input and output components
X = array[:,0:4]
Y = array[:,4]
binarizer = Binarizer(threshold=1.1).fit(X)
binaryX = binarizer.transform(X)
# summarize transformed data
np.set_printoptions(precision=3)
binaryX
```

Split data into Test and Training (30% to Test and 70% to Training)


```python
predictors = binaryX
targets = Y

TrainX, TestX, TrainY, TestY  =   train_test_split(predictors, targets, test_size=.3)
TrainX_scl = scaler.fit_transform(TrainX)
TestX_scl = scaler.transform(TestX)
```
